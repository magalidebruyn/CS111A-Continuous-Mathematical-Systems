```python
# Choosing starting distributions of visitors at our nodes

# Starting distribution 1 (v1)
v1 = vector([.6, .05, .2, .10, .05]) # 5 nodes: At Home (60%), Gym (5%), Yoga (20%), Sentido (10%), Heritage (5%)
# Starting distribution 2 (v2)
v2 = vector([.5, .05, .1, .3, .05]) # 5 nodes: At Home (50%), Gym (5%), Yoga (10%), Sentido (30%), Heritage (5%)
# Starting distribution 3 (v3)
v3 = vector([.8, 0, .05, .1, .05]) # 5 nodes: At Home (80%), Gym (0%), Yoga (5%), Sentido (10%), Heritage (5%)
# Starting distribution 4 (v4) : all students in the res
v4 = vector([1, 0, 0, 0, 0])  # 5 nodes: At Home (100%), Gym (0%), Yoga (0%), Sentido (0%), Heritage (0%)
```

```python
# Finding the stationary distribution
# by repeatedly mutliplying vectors by M

def find_stationary_dist(v, M):
    og = -1
    convergence = []
    for i in range(0,1000):
        value = (M**i)*v
        convergence.append((value))
        if value == og:
            print "The stationary distribution was found after", i, "multiplications by M."
            break
        og = value
    print "The stationary distribution is:"
    show (convergence[len(convergence)-1])
    return convergence

print "Finding v1's stationary distribution"
convergence1 = find_stationary_dist(v1, M)
stat_dist1 = convergence1[len(convergence1)-1]
print "\nFinding v2's stationary distribution"
convergence2 = find_stationary_dist(v2, M)
stat_dist2 = convergence2[len(convergence2)-1]
print "\nFinding v3's stationary distribution"
convergence3 = find_stationary_dist(v3, M)
stat_dist3 = convergence3[len(convergence3)-1]
print "\nFinding v4's stationary distribution"
convergence4 = find_stationary_dist(v4, M)
stat_dist4 = convergence4[len(convergence4)-1]
```

```python
# Checking resulting stationary distributions using alternative method

# For v1
convergence = []
for i in enumerate(range(1,150)):
    value = (M**i[1])*v1
    convergence.append((i[1], value))
    if i[0] != 0:
        if convergence[i[0]][1] == convergence[i[0]-1][1]:
            print('\nConvergence after '+str(i[1])+' multiplications' )
            print(convergence[-1][1])
            break

# For v2
convergence = []
for i in enumerate(range(1,150)):
    value = (M**i[1])*v2
    convergence.append((i[1], value))
    if i[0] != 0:
        if convergence[i[0]][1] == convergence[i[0]-1][1]:
            print('\nConvergence after '+str(i[1])+' multiplications' )
            print(convergence[-1][1])
            break

# For v3
convergence = []
for i in enumerate(range(1,150)):
    value = (M**i[1])*v3
    convergence.append((i[1], value))
    if i[0] != 0:
        if convergence[i[0]][1] == convergence[i[0]-1][1]:
            print('\nConvergence after '+str(i[1])+' multiplications' )
            print(convergence[-1][1])
            break
```

```python
# Display the results of convergence
import matplotlib.pyplot as plt
from sage.plot.scatter_plot import ScatterPlot

def graph_results(convergence, eigenvector):
    # start by adding eigenvector
    x = [len(convergence)+5]
    at_home = [eigenvector[0]]
    gym = [eigenvector[1]]
    yoga = [eigenvector[2]]
    sentido = [eigenvector[3]]
    heritage = [eigenvector[4]]
    # collect the data
    for i in range(len(convergence)):
        x.append(i)
        at_home.append(convergence[i][0])
        gym.append(convergence[i][1])
        yoga.append(convergence[i][2])
        sentido.append(convergence[i][3])
        heritage.append(convergence[i][4])
    # plot it
    plot = plt.subplot(1, 1, 1)
    plot.scatter(x, at_home, color = "red", label = "At Home")
    plot.scatter(x, gym, color = "black", label = "Gym")
    plot.scatter(x, yoga, color = "blue", label = "Yoga")
    plot.scatter(x, sentido, color = "orange", label = "Sentido")
    plot.scatter(x, heritage, color = "green", label = "Heritage")
    # plot.scatter([110, 110, 110, 110, 110], list(eigenvector), color = "pink", label = "Eigenvector")
    # add labels, legend
    plt.xlabel("Time passed (hours)")
    plt.ylabel("Distribution of Minervans (out of 1)")
    plt.title("Minerva students in Hyderabad")
    plot.legend()
    show(plot)
    plot.clear()

graph_results(convergence1, eigenvector)
graph_results(convergence2, eigenvector)
graph_results(convergence3, eigenvector)
graph_results(convergence4, eigenvector)
```
